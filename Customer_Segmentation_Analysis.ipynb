```python
from mpl_toolkits.mplot3d import Axes3D

sns.set_style("white")
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(custData.Age, custData["Annual Income (k$)"], custData["Spending Score (1-100)"], c='blue', s=60)
ax.view_init(30, 185)
plt.xlabel("Age",fontsize=14)
plt.ylabel("Annual Income (k$)", fontsize=14)
ax.set_zlabel('Spending Score (1-100)',fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
ax.set_zticklabels(labels=range(-20,120,20), fontsize=14)
plt.show()
```

```python
from sklearn.cluster import KMeans

wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(custData.iloc[:,1:])
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))    
plt.grid()
plt.plot(range(1,11),wcss, linewidth=2, color="red", marker ="8")
plt.xlabel("K Value",fontsize=14)
plt.xticks(np.arange(1,11,1), fontsize=14)
plt.ylabel("WCSS", fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

We can clearly notice that the elbow takes a bend at 5 so we can conclude that the number of clusters would be 5

```python
km = KMeans(n_clusters=5)
clusters = km.fit_predict(custData.iloc[:,1:])

custData["label"] = clusters

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
 

fig = plt.figure(figsize=(12,16));
ax = fig.add_subplot(111, projection='3d');
ax.scatter(custData["Age"][custData["label"] == 0], custData["Annual Income (k$)"][custData["label"] == 0], custData["Spending Score (1-100)"][custData["label"] == 0], c='blue', s=60);
ax.scatter(custData["Age"][custData["label"] == 1], custData["Annual Income (k$)"][custData["label"] == 1], custData["Spending Score (1-100)"][custData["label"] == 1], c='red', s=60);
ax.scatter(custData["Age"][custData["label"] == 2], custData["Annual Income (k$)"][custData["label"] == 2], custData["Spending Score (1-100)"][custData["label"] == 2], c='green', s=60);
ax.scatter(custData["Age"][custData["label"] == 3], custData["Annual Income (k$)"][custData["label"] == 3], custData["Spending Score (1-100)"][custData["label"] == 3], c='orange', s=60);
ax.scatter(custData["Age"][custData["label"] == 4], custData["Annual Income (k$)"][custData["label"] == 4], custData["Spending Score (1-100)"][custData["label"] == 4], c='purple', s=60);
ax.view_init(30, 185);
plt.xlabel("Age", fontsize=14);
plt.ylabel("Annual Income (k$)", fontsize=14);
ax.set_zlabel('Spending Score (1-100)',fontsize=14);
plt.xticks(fontsize=14);
plt.yticks(fontsize=14);
ax.set_zticklabels(labels=range(-20,120,20), fontsize=14);
plt.show();
```
